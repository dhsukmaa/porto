# Introduction
The occurrence of changes in humans can cause discomfort that can disrupt their mental stability. Just as a high school graduate goes on to college with a striking difference in culture, the effects of these changes can cause emotional and mental tension. In their journey, students will experience a disturbed mental phase which is influenced by several aspects such as majors, age, and year of study. Mental disorders that students can experience such as:

***Depression*** 
p2ptm the Indonesian Ministry of Health said that depression can be marked by a prolonged feeling of sadness and loss of interest in activities that we usually do happily. The next sign is to stop doing your usual daily activities for at least two weeks.The following are some of the psychological symptoms of someone who is depressed:

1.   Loss of interest or motivation to do something.
2.   Constantly feeling sad, even crying constantly.
3.   Feeling very guilty and worrying too much.
4.   Unable to enjoy life because of loss of self-confidence.
5.   Hard to make decisions and easily offended.

***Anxiety*** the Indonesian Ministry of Health said Anxiety disorder is a psychological condition when a person experiences excessive anxiety that is constant and difficult to control, resulting in a negative impact on his daily life.Meanwhile, physical symptoms that may accompany anxiety disorders include:


1.   Hard to sleep
2.   Body shaking
3.   Excessive sweating
4.   Muscles get tense
5.   Heart beat


***Panic Attack***  the Indonesian Ministry of Health said panic attack is a type of anxiety disorder characterized, by repeated 'panic attacks'.Panic attack symptoms include:

1.   Heart pounding
2.   Sweating
3.   Shaky
4.   Difficulty breathing
5.   Choking feeling

**Model Selection** We run model selection to get better prediction accuracy and make sure that our model is interpretable.

*   Dropping target
*   Assigning targets and features
*   Split data into trainig and testing sets
*   The model pipelines with preprocessing
*   Getting the confusion matrix and accuracy scores





```python
#Import Library

from sklearn.svm import SVC
from sklearn.svm import LinearSVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
from sklearn.metrics import precision_score, recall_score, confusion_matrix, classification_report, accuracy_score, f1_score
from sklearn import metrics
from sklearn.metrics import roc_curve, auc, roc_auc_score
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

import warnings
warnings.filterwarnings("ignore")
```

```python
#drop target 
X = df.drop(["CGPA"],axis=1)
y = df["CGPA"]
```

```python
#split test and train sets
X_train, X_test, y_train,y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
#A quick model selection process
#pipelines(a linear sequence of data preparation options, modeling operations, and prediction transform operations.)
#pipelines of models( it is short was to fit and pred)
pp_dt=Pipeline([ ('dt_classifier',DecisionTreeClassifier(random_state=42))])

pp_rf=Pipeline([('rf_classifier',RandomForestClassifier(random_state=0))])

pp_svc=Pipeline([('sv_classifier',SVC())])

# List of all the pipelines
pipelines = [pp_dt, pp_rf, pp_svc]

# Dictionary of pipelines and classifier types for ease of reference
pipe_dict = {0: 'Decision Tree', 1: 'RandomForest', 2: "SVC"}
```

```python
# Fit the pipelines
for pipe in pipelines:
    pipe.fit(X_train, y_train)

#cross validation on accuracy 
cv_results_accuracy = []
for i, model in enumerate(pipelines):
    cv_score = cross_val_score(model, X_train,y_train, cv=10 )
    cv_results_accuracy.append(cv_score)
    print("%s: %f " % (pipe_dict[i], cv_score.mean()))
```

**what we get**

Random Forest to be the most accurate.Let's build a better random forest with grid search cv. 

**Model Evaluation**

*   Build a Random Forest model.
*   Create classification .
*   Visualize the confusion matrix


```python
#taking look at the test set
pred_rfc = pp_rf.predict(X_test)
accuracy = accuracy_score(y_test, pred_rfc)
print(accuracy)
```

```python
RF_model = RandomForestClassifier()
RF_model.fit(X_train, y_train)
#Testing the Model on test set
predictions=RF_model.predict(X_test)
acc1= accuracy_score(y_test,predictions)
acc1
```

```python
acc2 = accuracy_score(y_test, predictions)
recall = recall_score(y_test, predictions, average="weighted")
precision = precision_score(y_test, predictions, average="weighted")
f1_score = f1_score(y_test, predictions, average="micro")

print("Random Forest Results ")
print('\n')
print("Accuracy    : ", acc2)
print("Recall      : ", recall)
print("Precision   : ", precision)
print("F1 Score    : ", f1_score)
```

```python
print(classification_report(y_test, predictions))
```
